```python
df["Review Rating"].fillna(df["Review Rating"].median(),inplace=True)
```

```python
df.isnull().sum()
```

```python
columns=df.columns.to_frame().rename(columns={0:"Column_Name"})
columns
```

```python
columns["Column_Name"]=columns["Column_Name"].str.replace(" ","_").str.lower().to_frame()
columns
```

```python
df.columns=columns["Column_Name"]
df.rename(columns={"purchase_amount_(usd)":"purchase_amount"},inplace=True)
df.columns.to_list()
```

```python
# create a column age_group
condition=[df["age"]>=60,df["age"]>=38,df["age"]>=25]
choice=["Senior","Middle-aged","Adult"]
df["age_group"]=np.select(condition,choice,default="Young Adult")
df
```

```python
# create columns purchase_frquency_days

frequency_mapping={
    "Fortnightly":14,
    "Weekly":7,
    "Monthly":30,
    "Quarterly":90,
    "Bi-Weekly":14,
    "Annually":365,
    "Every 3 Months":90
}
df["purchase_frequency_days"]=df["frequency_of_purchases"].map(lambda x: frequency_mapping[x])
```
